# Data Visualization

```python
# import libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import titanic data form seaborn library
df =sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare', color='red')
```

```python
# change colors using hex codes
df.plot(kind='scatter', x='age', y='fare', color='#0E93EC')
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
plt.ylabel('age')
plt.xlabel('gender')
```

```python
# seaborn plot 

# bar plot using seaborn on df 
sns.barplot(data =df, x='survived', y='fare')
```

```python
sns.barplot(data =df, x ='sex' ,y='fare' , hue='class')
```

```python
sns.barplot(data =df, x ='sex' ,y='age' , hue='class')
```

```python
sns.boxplot(data =df, x ='sex' ,y='age' , hue='class')
```

```python
sns.scatterplot(data =df, x ='sex' ,y='age' , hue='class')
```

```python
sns.scatterplot(data =df, x ='fare' ,y='age' , hue='class')
```

```python
sns.scatterplot(data =df, y ='fare' ,x='age' , hue='class')
```

```python
# makses scatter plot used ploty 
import plotly.express as px 
fig =px.scatter(df, x='age', y='fare', color='class')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title='Mt Bruno Elevation', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```

```python
import plotly.express as px
df = px.data.gapminder()
px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country",
           log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```
